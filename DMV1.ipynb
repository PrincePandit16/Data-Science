```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('dmvprac1_sales_data_sample.csv',encoding='ISO-8859-1')
df2 = pd.read_excel('sales_excel.xlsx')
df3 = pd.read_json('sales_json.json')
```

```python
combined_data = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
Q1 = np.percentile(combined_data['SALES'],25,interpolation = 'midpoint')
Q3 = np.percentile(combined_data['SALES'],75,interpolation = 'midpoint')
IQR = Q3-Q1
print("Old Shape: ",combined_data.shape)
upper = np.where(combined_data['SALES']>=(Q3+1.5*IQR))
lower = np.where(combined_data['SALES']<=(Q1-1.5*IQR))
combined_data.drop(upper[0],inplace=True)
combined_data.drop(lower[0],inplace=True)
print("New Shape: ",combined_data.shape)
sns.boxplot(x='SALES',data=combined_data)
```
